```python
import pygame
import sys

# Initialisatie van Pygame
pygame.init()

# Scherminstellingen
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("2D Platform Game")

# Kleuren
white = (255, 255, 255)
blue = (0, 128, 255)

# Karakterinstellingen
player_width = 20
player_height = 70
player_x = (screen_width - player_width) // 2
player_y = screen_height - player_height - 10
player_speed = 5
jump_height = 15  # Pas dit aan voor de gewenste springhoogte
is_jumping = False
jump_count = 10

# Platforminstellingen
platform_width = 100
platform_height = 10
platform_color = (128, 128, 128)
platforms = [(150, 500), (300, 400), (500, 300)]  # Lijst van platformposities

# Vloerinstellingen
floor_y = screen_height - 10  # 10 pixels boven de onderkant van het scherm

# Hoofdloop
clock = pygame.time.Clock()

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_x > 0:
        player_x -= player_speed
    if keys[pygame.K_RIGHT] and player_x < screen_width - player_width:
        player_x += player_speed

    # Springen
    if not is_jumping:
        if keys[pygame.K_SPACE] and (player_y == floor_y - player_height or any(platform_x < player_x < platform_x + platform_width and platform_y - player_height < player_y < platform_y for platform_x, platform_y in platforms)):
            is_jumping = True
    else:
        if jump_count >= -10:
            neg = 1
            if jump_count < 0:
                neg = -1
            player_y -= (jump_count ** 2) * 0.5 * neg
            jump_count -= 1
        else:
            is_jumping = False
            jump_count = 10

    # Zwaartekracht simulatie
    if player_y < floor_y - player_height:
        on_platform = False
        for platform_x, platform_y in platforms:
            if platform_x < player_x < platform_x + platform_width and platform_y - player_height < player_y < platform_y:
                on_platform = True
                player_y = platform_y - player_height
                break
        if not on_platform:
            player_y += 5  # Pas dit aan voor de gewenste zwaartekrachtsnelheid
    else:
        player_y = floor_y - player_height  # Zorg ervoor dat het personage op de vloer blijft

    # Update het scherm
    screen.fill(white)

    # Tekenen van de platforms
    for platform in platforms:
        pygame.draw.rect(screen, platform_color, (platform[0], platform[1], platform_width, platform_height))

    # Tekenen van de stickfigure
    pygame.draw.line(screen, blue, (player_x + player_width // 2, player_y),
                     (player_x + player_width // 2, player_y + player_height), 5)  # Body
    pygame.draw.line(screen, blue, (player_x + player_width // 2, player_y + player_height),
                     (player_x, player_y + player_height + 20), 5)  # Left leg
    pygame.draw.line(screen, blue, (player_x + player_width // 2, player_y + player_height),
                     (player_x + player_width, player_y + player_height + 20), 5)  # Right leg
    pygame.draw.circle(screen, blue, (player_x + player_width // 2, player_y), 10)  # Head

    pygame.display.flip()

    # Beperk framesnelheid
    clock.tick(60)
```
